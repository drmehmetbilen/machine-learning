```python
import pandas as pd 

data = pd.read_csv("../data/iris.csv", names=["sl","sw","pl","pw","class"])
data.head()
```

```python
sample_1 = [4.8,3.0,1.4,0.1] # setosa
sample_2 = [6.4,2.9,4.3,1.3] # versicolor
sample_3 = [6.3,2.5,5.0,1.9] # virgina

K = 3
```

```python
def calculate_distance(ins1, ins2):
    sum = 0
    for index in range(len(ins1)):
        sum += (ins1[index] - ins2[index])**2
    sqr = sum**0.5
    return sqr
```

```python
#test distance function
sample_a = [1,1,2,3,5]
sample_b = [2,2,2,3,4]
calculate_distance(sample_a,sample_b)
```

```python
print("To Classify : ",sample_2)
distance_list = []
for index, row in data.iterrows():
    data_sample = row.values[:-1]
    true_class = row.values[-1]
    distance = calculate_distance(data_sample, sample_2)
    print(data_sample, true_class, distance)
    distance_list.append([true_class, distance])
```

```python
sorted_distance_list = sorted(distance_list, key = lambda x: x[1])
sorted_distance_list
```
